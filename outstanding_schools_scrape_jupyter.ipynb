```python
# Program that obtains the number of 

# import the necessary libraries
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://snobe.co.uk/ofsted-outstanding-schools/london'
source = requests.get(url).text
soup = BeautifulSoup(source, 'lxml')

# obtain the number of pages of outstanding schools
num_pages = int(soup.find("li", class_='page-item right_corner').a.text)
```

```python
#Obtain all of the URL's for all of the outstanding schools
school_links = []
counter = 1

while counter <= num_pages:
	if counter == 1:
		source = requests.get(url).text
	else:
		source = requests.get(url+f"/page-{counter}").text

	soup = BeautifulSoup(source, 'lxml')
	b = soup.find_all("h2", class_="school_list_title")
	for i in b:
		school_links.append(i.a['href'])

	counter += 1
```

```python
school_links
```

```python
outstanding_locations = []
for school in school_links:
    source = requests.get(school).text
    soup = BeautifulSoup(source, 'lxml')

    a = soup.find_all('script')
    b = str(a[-9])
    c = b.split(";")
    lat_lon_string = []
    for item in c:
        if ('vlatitude' in item) or ('vlongitude' in item):
            lat_lon_string.append(item)

    lats_lons = []
    d = lat_lon_string[-2:]
    for i in d:
        lats_lons.append(i[-12:-1])
    outstanding_locations.append(lats_lons)
```

```python
container = -1
for location in outstanding_locations:
    container += 1
    for i in location:
        if "'" in i:
            new_str = i.replace("'", "")
            outstanding_locations[container][1] = new_str
```
